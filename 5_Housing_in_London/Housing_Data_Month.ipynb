```python
import pandas as pd
import numpy as np
```

```python
# Date is from 1995 to 2019
```

```python
house_month = pd.read_csv('housing_in_london_monthly_variables.csv')
house_month.head()
```

```python
house_month.shape
```

```python
house_month.isnull().mean()*100
```

```python
house_month.drop('no_of_crimes', axis = 1, inplace = True)
```

```python
house_month.dropna(axis=0, subset= ['houses_sold'], inplace= True)
house_month.shape
```

```python
house_month.isnull().mean()*100
```

```python
house_month.info()
```

```python
house_month['date'] = pd.to_datetime(house_month['date'])
```

```python
house_month.info()
```

```python
# house_month['total_cost'] = house_month['average_price'] * house_month['houses_sold']
# house_month.head()
```

```python
house_m_group_2 = house_month.groupby(['code', 'area'])['average_price', 'houses_sold', 'borough_flag'].mean()
house_m_group_2 = house_m_group_2.reset_index()
house_m_group_2
```

```python
house_m_group_3 = house_month.groupby( house_month['date'].dt.to_period('Y'))['average_price', 'houses_sold'].mean()
house_m_group_3 = house_m_group_3.reset_index()
house_m_group_3
```

```python
house_m_group_4 = house_month.groupby(['area', house_month['date'].dt.to_period('Y')])['average_price', 'houses_sold'].mean()
house_m_group_4 = house_m_group_4.reset_index()
house_m_group_4
```
