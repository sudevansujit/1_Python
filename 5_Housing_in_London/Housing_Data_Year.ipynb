```python
import pandas as pd
import numpy as np
```

```python
house_year = pd.read_csv('housing_in_london_yearly_variables.csv')
house_year.head()
```

```python
house_year.shape
```

```python
house_year['borough_flag'].value_counts()
```

```python
house_year.info()
```

```python
house_year.drop(['life_satisfaction', 'area_size', 'no_of_houses'], axis = 1, inplace = True )
house_year.info()
```

```python
house_year.dropna(axis = 0, subset = ['recycling_pct'], inplace = True)
house_year.info()
```

```python
house_year.dropna(axis = 0, subset = ['number_of_jobs'], inplace = True)
house_year.info()
```

```python
house_year.dropna(axis = 0, subset = ['median_salary'], inplace = True)
house_year.info()
```

```python
house_year.dropna(axis = 0, subset = ['population_size'], inplace = True)
house_year.info()
```

```python
house_year['borough_flag'].value_counts()
```

```python
down = house_year[house_year['mean_salary'] == '#']
down
```

```python
down.index
```

```python
house_year.drop(down.index, axis = 0,  inplace = True)
```

```python
house_year.shape
```

```python
house_year['mean_salary'] = house_year['mean_salary'].astype(dtype = float) 
```

```python
house_year['recycling_pct'] = house_year['recycling_pct'].astype(dtype = int) 
```

```python
house_year['date'] = pd.to_datetime(house_year['date'])
```

```python
house_year.info()
```

```python
house_group_2 = house_year.groupby(['code','area'])['median_salary', 'mean_salary', 'population_size', 'recycling_pct','borough_flag' 
                                           ,'number_of_jobs'].mean().apply(lambda x: round(x, 2))
house_group_2 = house_group_2.reset_index()
house_group_2
```

```python
house_group_3 = house_year.groupby(house_year['date'].dt.to_period('Y'))['median_salary', 'mean_salary', 'population_size', 'recycling_pct' 
                                           ,'number_of_jobs'].mean().apply(lambda x: round(x, 2))
house_group_3 = house_group_3.reset_index()
house_group_3
```

```python
house_group_4 = house_year.groupby(['area', house_year['date'].dt.to_period('Y')])['median_salary', 'mean_salary', 'population_size', 'recycling_pct' 
                                           ,'number_of_jobs'].mean().apply(lambda x: round(x, 2))
house_group_4 = house_group_4.reset_index()
house_group_4
```
